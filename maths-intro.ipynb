# Multivariable Functions
* If input is made up of multiple variables
$$
f(x,y) = x^{2}y 
$$
* If the output of a function consists of multiple numbers, which are commonly called as **vector valued** functions 

$$
f(x) = \begin{bmatrix} cos(x) \\ sin(x) \end{bmatrix}
$$

* $f(x,y) = x^{2}y $ we can associate this as a map from 2D plane to points on the number line

* To predict a user's behavior, it might create a function that takes in thousands of variables, including the user's age, the coordinates of their location, the number of times they've clicked on links of a certain type, etc. The output might also include multiple variables, such as the probability they will click on a different link or the probability they purchase a different item


## Scalar Valued Function

$$ f(x,y) = x*e^{x^{2} - y^{2}}$$


```python
x,y = np.meshgrid(np.linspace(-5,5,100), np.linspace(-5,5,100))
z = np.exp(-x**2 - y**2)
fig = plt.figure()
au = fig.add_subplot(111, projection='3d')
au.plot_surface(x,y,z)
plt.show()
```

### Contour Maps
$$
f(x,y) = x^{4} - x^{2} + y^{2}
$$

```python
i,j = np.meshgrid(np.linspace(-2,2,1000), np.linspace(-2,2,1000))
f =  -(i**4 - i**2 + j**2)
fig = plt.figure()
au = fig.add_subplot(111)
au.contour(i,j,f)

plt.show()
```

## Vector Valued Function

### Parametric Curves 
$$
f(x) = \begin{bmatrix}
        cos(x) \\
        x  sin(x)
        \end{bmatrix}
$$

```python
t = np.linspace(0,12,100)
x = np.cos(t)
y = t*np.sin(t)
fig = plt.figure()
ay = fig.add_subplot(111)
ay.plot(x,y)
ay.set_xlim(-5,5)
ay.set_title('parametric curve : time to location')
plt.show()
```

#### Parametric Surface
A function with two dimensional input and three dimensional output 

$$
f(t,s) = \begin{bmatrix}
t^{3} - st \\
s - t \\
s + t 
\end{bmatrix}
$$
such that
 
$ 0< s <3 $ ,  $-2<t<2$

```python
s = np.linspace(0,3,100)
t = np.linspace(-2,2,100)
fig2 = plt.figure()
ac = fig2.add_subplot(111, projection='3d')
S,T = np.meshgrid(s,t)
x = T**3 - S*T 
y = S - T
z = S + T
ac.plot_surface(x,y,z)
plt.show()
```

#### Parametrizing a surface

Starting with surface and finding the function that created the surface is know as parametrizing the surface

### 3d vector fields 
$$ 
f(x,y,z) = \begin{bmatrix} x \\ y \\ z \end{bmatrix}
$$

```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
z = np.linspace(-5,5,10)
X,Y,Z = np.meshgrid(x,y,z)
U = X
V = Y
W = Z
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X,Y,Z,U,V,W, length=0.1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

To analyze the function using components.

```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
z = np.linspace(-5,5,10)
X,Y,Z = np.meshgrid(x,y,z)
U = Y*Z
V = X*Z
W = X*Y
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X,Y,Z,U,V,W, length=0.1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

### Transformations

We can see functions as a transformations from input space to output space 
$$
f(x) = x^{2} - 3 \; 
$$
where $ x \in \mathbb{R}$

$$
f(x) = \begin{bmatrix}
        cos(x) \\
        x  sin(x)
        \end{bmatrix}
$$

```python
t = np.linspace(0,12,100)
x = np.cos(t)
y = t*np.sin(t)
fig = plt.figure()
ay = fig.add_subplot(111)
ay.plot(x,y)
ay.set_xlim(-5,5)
plt.show()
```

```python

t,s = np.meshgrid(np.linspace(0,2*np.pi),np.linspace(0,2*np.pi))
x = 3*np.cos(t) + np.cos(t)*np.cos(s)
y = 3*np.sin(t) + np.sin(t)*np.cos(s)
z = np.sin(s)
fig = plt.figure()
az = fig.add_subplot(111, projection='3d')
az.plot_surface(x,y,z)
az.set_aspect('equal')
plt.show()
```
